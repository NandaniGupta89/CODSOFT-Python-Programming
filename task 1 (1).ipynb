# Task 1 - To-Do List


```python
#To-do list

from prettytable import PrettyTable

print("\nMy To-Do List.")
task = "\n1: Enter 1 to create a new in To-Do List.\n2: Enter 2 to delete from To-Do List.\n3: Enter 3 to update in To-Do List.\n4: Enter 4 to track your To-Do List.\n5: Enter for exit."
print(task)

list = []
x = PrettyTable()

def my_list():
    x.field_names = ["Item names"]
    for i in list:
        x.add_row([i])
    print(x.get_string(title="To-Do List"))
    x.clear_rows()
run = True
while run:
    user_input = input("\nWhat do you want to do?(1,2,3,4,5): ")
    if user_input == '1':
        new = input("\nPlease enter your new To-Do: ").lower()
        print(f"\nYour current To-Do is {new}.")
        list.append(new)
    elif user_input == '2':
        while True:
            item_name = input("\nPlease enter a name of an item you want to delete.").lower()
            try:
                if item_name in list:
                    choice = input(f"Are you sure to delete {item_name} from your To-DO List? (Y/N)").lower()
                    if choice == "y":
                        list.remove(item_name)
                        print("Your updated To-Do List.")
                        my_list()
                        break
                else:
                    print("Item not found.")
            except Exception:
                print("Something went wrong.")
    elif user_input == '3':
        while True:
            item_name = input("\nPlease enter a name of an item you want to update.").lower()
            try:
                if item_name in list:
                    choice = input(f"Are you sure to update {item_name} from your To-DO List? (Y/N)").lower()
                    if choice == "y":
                        update_item = input(f"\nPlease enter a name you want to update {item_name} with ").lower()
                        index = list.index(item_name)
                        list[index] = update_item
                        print("Your updated To-Do List.")
                        my_list()
                        break
                else:
                    print("Item not found.")
                    break
            except Exception:
                print("Something went wrong.")
    elif user_input == '4':
        my_list()
    elif user_input == '5':
        ask_user = input("\nAre you sure to exit?(Y/N):").lower()
        if ask_user == "y":
            run = False
    elif user_input == "" or user_input == " ":
        print("Please enter something.")
    else:
        print("Please enter a valid value.")
    
```
